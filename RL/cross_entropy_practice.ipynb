```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random

class Net(nn.Module):  
    def __init__(self, input_size, output_size, hidden_neurons=20):
        super(Net, self).__init__()
        self.input_size = input_size
        self.output_size = output_size
        self.fc1 = nn.Linear(input_size, hidden_neurons)
#         self.bn1 = nn.BatchNorm1d(hidden_neurons)
        self.fc2 = nn.Linear(hidden_neurons, hidden_neurons)
        self.fc3 = nn.Linear(hidden_neurons, output_size)
        self.criterion = nn.CrossEntropyLoss()
        self.optimizer = optim.Adam(self.parameters(), lr=0.001)
        
    def forward(self, x):
        x = self.fc1(x)
#         x = self.bn1(x)
        x = F.tanh(x)
        x = self.fc2(x)
        x = F.tanh(x)
        x = self.fc3(x)
        return x
    
    def inference(self, x):
        x = self.forward(x)
        return F.softmax(x, dim=-1)[0]
```

```python
class NeuralAgent:
    def __init__(self, input_size, output_size, epsilon=0.01):
        self.net = Net(input_size, output_size)
        self.net.train(False)

    def get_action(self, observation, epsilon):
        if np.random.uniform() < epsilon:
            return env.action_space.sample()
        net_input = Variable(torch.Tensor(observation).unsqueeze(0))
        output_probas = self.net.inference(net_input).data
        predicted_action = np.random.multinomial(1, output_probas).argmax()
#         print(predicted_action, end='')
        return predicted_action

    def train(self, train_features, train_labels, 
              net_epoches_count=1, batch_size=1000):
        self.net.train(True)
        for net_epoch in range(net_epoches_count):
            order = np.random.permutation(train_features.shape[0])

            for start_index in range(0, train_features.shape[0], batch_size):
                self.net.optimizer.zero_grad()

                batch_indice = order[start_index:start_index+batch_size]

                batch_features = torch.FloatTensor(train_features[batch_indice, :])
                batch_labels = torch.LongTensor(train_labels[batch_indice])

                if batch_features.size()[0] == 1:  # batch norm fails
                    continue

                features_var = Variable(batch_features)
                targets_var = Variable(batch_labels)

                preds_var = self.net.forward(features_var)

                loss = self.net.criterion(preds_var, targets_var)
                loss.backward()

                self.net.optimizer.step()
        self.net.train(False)

from sklearn.linear_model import LogisticRegression
class LRAgent:
    def __init__(self):
        self.lr = LogisticRegression()

    def get_action(self, observation, epsilon):
        if np.random.uniform() < epsilon:
            return env.action_space.sample()
        output_probas = self.lr.predict_proba(np.expand_dims(observation, axis=0))[0]
        predicted_action = np.random.multinomial(1, output_probas).argmax()
#         print(predicted_action, end='')
        return predicted_action
    def train(self, observations, actions):
        self.lr.fit(observations, actions)
```

```python
best_games_ratio = 0.2
games_count = 100
timestamps_count = 1000
epoches_count = 100

observation = env.reset()
agent = NeuralAgent(env.observation_space.shape[0], env.action_space.n)

mean_epoch_rewards = []

for epoch in range(epoches_count):
    epoch_rewards = []
    game_process2game_reward = []
    for game in range(games_count):
        env.reset()
        game_process = []
        game_reward = 0
        for timestamp in range(timestamps_count):
#             action = env.action_space.sample()
            if epoch == 0:
                action = env.action_space.sample()
            else:
                action = agent.get_action(observation, epsilon=(1 - epoch/epoches_count))
            observation, reward, done, info = env.step(action)
            game_reward += reward
            game_process.append((observation, action))
#             print(action, end='')
            if done:
                break
        epoch_rewards.append(game_reward)
                
        game_process2game_reward.append([game_process, game_reward])
    games_sorted_by_reward = sorted(game_process2game_reward, key=lambda x: -x[1])
    best_games = games_sorted_by_reward[:int(best_games_ratio * games_count)]
#     plt.hist(epoch_rewards)
#     plt.show()
    print(np.mean(epoch_rewards), end=' ')#, np.max(epoch_rewards), np.mean([r for p, r in best_games]))
    mean_epoch_rewards.append(np.mean(epoch_rewards))
    observations = []
    actions = []
    
    for game_process, game_reward in best_games:
        for observation, action in game_process:
            observations.append(observation)
            actions.append(action)
    
    agent.train(np.array(observations), np.array(actions))

plt.plot(mean_epoch_rewards)
```
