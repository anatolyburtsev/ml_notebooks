```python
import pandas as pd
import folium
import numpy as np

from folium.plugins import HeatMap
```

```python
data = pd.read_csv("~/PycharmProjects/mts-geohack/data/zones.csv")
```

```python
data.head(3)
```

```python
heat_data = data[["lat_c", "lon_c", "calls_daily"]][data["calls_daily"] > 1].values.tolist()
```

```python
len(heat_data)
```

```python
folium_map = folium.Map(location=[55.8,37.3], tiles='Stamen Terrain', zoom_start=6)

# folium.CircleMarker(location=fp, fill=True).add_to(folium_map)
# folium_map
```

```python
HeatMap(heat_data, radius=5, blur=5).add_to(folium_map)
folium_map
```
